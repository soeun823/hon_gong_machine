**첫 번째 머신러닝 프로그램**<br>
:K-최근접 이웃 알고리즘 사용 => 도미와 빙어 데이터 구분

```python
lenght = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l,w] for l,w in zip(lenght,weight)] #첫번째 열: 길이, 두번째 열: 무게
# print(fish_data)
```

```python
fish_target = [1] * 35 + [0]*14 #1:도미, 2:빙어
# print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier #k-최근접 이웃알고리즘 라이즈러리
kn = KNeighborsClassifier()

kn.fit(fish_data, fish_target) #도미를 찾기위한 기준을 학습시킴
kn.score(fish_data,fish_target)
```

```python
kn.predict([[30,600]]) #새로운 물고기 데이터 30cm,600g
```

```python
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49) #참고 데이테를 49개로 한 kn49모델

kn49.fit(fish_data, fish_target)
kn49.score(fish_data,fish_target)
# print(35/49)
#결론 : n-neighbor 매개변수를 49로 두는것은 별로다.
```
